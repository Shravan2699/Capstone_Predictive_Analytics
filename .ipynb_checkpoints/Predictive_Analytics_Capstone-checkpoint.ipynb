### *Business Understanding*

The primary goal of this analysis is to predict life expectancy based on government healthcare expenditures, critical illness rates, and temporal data. Life expectancy is a critical metric for assessing the overall health and well-being of a population and serves as an essential indicator for policy-makers and public health officials.Governments allocate substantial funds to healthcare services, with the expectation that higher spending improves healthcare infrastructure, reduces disease burden, and ultimately enhances life expectancy.

### *Problem Statement*

1. To what extent do government healthcare expenditures correlate with life expectancy?
2. Can we predict life expectancy based on healthcare expenditure and critical illness rates?
3. What are the most important features for predicting life expectancy?

## **Data Preparation**

### *1. Handling Missing/Null Value*

```python
# Checking if there is null- values
df.isnull().sum()
```

```python
df.info()
```

&#128161;Based on the two tables, there are 203 missing values in Expenditure by Provincial/Territorial Government and 234 missing values in Expenditure by Public Government.We also have 1 missing value in Critical Illness rate (per 10,000 population)

```python
# Imputation of features Expenditure by Provincial/Territorial Government and Expenditure by Public Government and Critical Illness rate (per 10,000 population)
# Impute missing values in 'Expenditure by Provincial/Territorial Government' , 'Expenditure by Public Government' and  'Critical Illness rate (per 10,000 population)' with the median
# Median imputation without using inplace
df['Expenditure by Provincial/Territorial Government'] = df['Expenditure by Provincial/Territorial Government'].fillna(df['Expenditure by Provincial/Territorial Government'].median())
df['Expenditure by Public Government'] = df['Expenditure by Public Government'].fillna(df['Expenditure by Public Government'].median())
df['Critical Illness rate (per 10,000 population)'] = df['Critical Illness rate (per 10,000 population)'].fillna(df['Critical Illness rate (per 10,000 population)'].median())
```

```python
# Let's Inspect the dataset
df.info()
```

To impute missing values in the "Expenditure by Provincial/Territorial Government" and 'Expenditure by Public Government' and 'Critical Illness rate (per 10,000 population)' features, using the median is recommended due to its robustness to outliers and ability to preserve the right-skewed distribution

### *3. Feature Selection*

### 3.1. Feature Importance Analysis

```python
import pandas as pd
from sklearn.ensemble import ExtraTreesRegressor
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Define the target variable and features
y = df['Life Expectancy']

# Drop the target and unnecessary columns for feature selection
X = df.drop(['Life Expectancy', 'Year'], axis=1)

# Convert categorical columns to numerical using Label Encoding
categorical_cols = X.select_dtypes(include=['object']).columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    X[col] = label_encoder.fit_transform(X[col])

# Fill missing values using the median for numerical columns
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Train ExtraTreesRegressor on the preprocessed data
model = ExtraTreesRegressor()
model.fit(X_imputed, y)

# Extract feature importances
feature_importances = pd.Series(model.feature_importances_, index=X.columns)

# Plot the top 10 feature importances
top_features = feature_importances.nlargest(10)
top_features.plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()
```

&#128161;The plot shows the importance of different features in a machine learning model. "Critical Illness Rate" is the most important feature, followed by "Province". Other features like "Use of Funds" and "Total Expenditure" have less impact on the model's predictions.

### 3.2. Chi-Squared

```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Define the target variable and features
y = df['Life Expectancy']

# Drop the target and unnecessary columns for feature selection
X = df.drop(['Life Expectancy', 'Year'], axis=1)

# Convert categorical columns to numerical using Label Encoding
categorical_cols = X.select_dtypes(include=['object']).columns
label_encoder = LabelEncoder()
for col in categorical_cols:
    X[col] = label_encoder.fit_transform(X[col])

# Fill missing values using the median for numerical columns
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# Use f_regression for continuous target variable
bestfeatures = SelectKBest(score_func=f_regression, k='all')
fit = bestfeatures.fit(X_imputed, y)

# Create a DataFrame for feature scores
dfscores = pd.DataFrame(fit.scores_, columns=['Score'])
dfcolumns = pd.DataFrame(X.columns, columns=['Specs'])

# Concatenate the scores with column names
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores_sorted = featureScores.nlargest(10, 'Score')

# Display the top 10 features
print(featureScores_sorted)
```

## **CREATING MODELS**

## *1. Multiple Linear Regression*

```python
imputer = SimpleImputer(strategy='mean')  #We will impute the missing values with mean
X_imputed = imputer.fit_transform(X)

# We split the data into train and test data
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Fit the linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Calculate the predicted values
y_pred = regressor.predict(X_test)

# Use Mean Squared Error and R-square to check how the model performs
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Multiple Linear Regression")
print(f"MSE: {mse}\nR-squared: {r2}")
```

&#128161;The multiple linear regression model shows promising results. The R-squared value of 0.958 indicates that the model explains a significant portion of the variance in life expectancy, suggesting strong predictive accuracy.

```python

# Scatter plot of actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label='Predicted vs Actual')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2, label='Ideal Fit')
plt.xlabel('Actual Life Expectancy')
plt.ylabel('Predicted Life Expectancy')
plt.title('Predicted vs Actual Life Expectancy')
plt.legend()
plt.show()
```

&#128161;The plot shows a good agreement between the predicted and actual life expectancy values. The blue dots, representing the actual vs. predicted values, are clustered closely around the red line, indicating accurate predictions.

## *2. Random Forest Regressor*

```python
# Import Required Libraries
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_squared_error, r2_score
import numpy as np

# Initialize the Random Forest model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Set up k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Define scoring metrics
scoring = {
    'mse': make_scorer(mean_squared_error, greater_is_better=False),
    'r2': make_scorer(r2_score)
}

# Perform cross-validation
mse_scores = cross_val_score(rf, X, y, cv=kf, scoring=scoring['mse'])
r2_scores = cross_val_score(rf, X, y, cv=kf, scoring=scoring['r2'])

# Calculate average scores
avg_mse = -np.mean(mse_scores)  # Convert negative MSE to positive
avg_r2 = np.mean(r2_scores)

# Print results
print(f"Cross-Validated Mean Squared Error: {avg_mse:.4f}")
print(f"Cross-Validated R² Score: {avg_r2:.4f}")

# Train the model on the entire dataset (optional, for final use)
rf.fit(X, y)

# Feature Importance
importances = rf.feature_importances_
feature_names = X.columns

print("\nFeature Importance:")
for feature, importance in zip(feature_names, importances):
    print(f"{feature}: {importance:.4f}")
```

&#128161;The model uses a Random Forest Regressor to predict life expectancy. Cross-validation was used to assess the model's performance, resulting in a mean squared error of 0.0119 and an R-squared of 0.9984. This indicates a highly accurate model with strong predictive power. The feature importance analysis highlights "Province_Nunavut" as the most influential factor in the model's predictions.

```python
# Bar chart for feature importance
plt.figure(figsize=(8, 6))
plt.barh(feature_names, importances, color='skyblue')
plt.xlabel('Feature Importance')
plt.ylabel('Features')
plt.title('Feature Importance in Random Forest')
plt.show()
```

The plot shows the relative importance of different features in predicting life expectancy. "Province_Nunavut" has the highest importance, suggesting it has the strongest influence on the model's predictions. Other features like "Critical Illness Rate" and "Total Expenditure" also contribute significantly, while the impact of provincial features varies.
